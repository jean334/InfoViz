```python
def generate_all_month_list():
    first_month = game_dataset3['date'].min()
    last_month = game_dataset3['date'].max()

    month_list = []

    first_year = int(first_month.split('/')[0])
    first_month = int(first_month.split('/')[1])
    last_year = int(last_month.split('/')[0])
    last_month = int(last_month.split('/')[1])
    for year in range(first_year, last_year + 1):
        for month in range(1, 13):
            month_list.append(f"{year}/{month:02d}")
    return month_list

month_list = generate_all_month_list()
print("month list"  , month_list)
```
