```python
import numpy as np
import pandas as pd
import re
```

```python
data_engine = pd.read_csv('playtracker_data_engine.csv')
data_engine2 = data_engine.drop(["game_id", "active_players", "avg_total_playtime", "avg_recent_playtime", "median_recent_playtime", "release_date", "genre", "porting", "publisher", "mode", "perspective", "theme", "more_tags"], axis=1)
```

```python
data_engine2
def filterEmptyEngine(df):
    empty_indices = df[df['engine'] == '[]'].index
    df = df.drop(empty_indices)
    return df

data_engine3 = filterEmptyEngine(data_engine2)
```

```python
data_engine3['engine'].head(20)
#process data_engine3['engine'] into a proper list
def process_engine_column(engine_column):
    processed_list = []
    for item in engine_column:
        # Remove square brackets and split by comma
        item = re.sub(r'[\[\]]', '', item)
        item = re.sub(r'\'', '', item)
        item = re.sub(r'\s+', ' ', item)  # Remove extra spaces
        item = item.split(',')
        processed_list.append([x.strip() for x in item])
    return processed_list

data_engine3['engine'] = process_engine_column(data_engine3['engine'])
```

```python
engine_list = data_engine3['engine'].iloc[:].tolist()
#flatten the list of lists into a single list
engine_list_flat = [item for sublist in engine_list for item in sublist]
engine_list_flat
```

```python
def processEngine(value):
        if re.search("Unreal Engine [0-9]", value):
            return "Unreal Engine"
        elif re.search("Source [0-9]", value):
            return "Source"
        elif re.search("Halo .*", value):
            return "Halo"
        elif re.search("id Tech [0-9]", value):
            return "id Tech"
        elif re.search("Unity [0-9]", value):
            return "Unity"
        elif re.search("LithTech .*", value):
            return "LithTech"
        elif re.search("CryEngine [0-9]", value):
            return "CryEngine"
        elif re.search("Glacier [0-9]", value):
            return "Glacier"
        elif re.search("RPG Maker .*", value):
            return "RPG Maker"
        else:
            return value

for i,elem in enumerate(engine_list_flat):
    processed_value = processEngine(elem)
    engine_list_flat[i] = processed_value

engine_list_flat
```

```python
s_engines = pd.Series(engine_list_flat).unique()
s_count = pd.Series(engine_list_flat).value_counts()
#pd.DataFrame({'email':sf.index, 'list':sf.values})

#df_engine
#engine_list_flat
#engine_list_flat.to_csv('./my-app/public/assets/engines_hist.csv', index=True, header=False)
```
