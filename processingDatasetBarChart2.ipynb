```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import re 
import json
```

```python
data_engine = pd.read_csv('playtracker_data_engine.csv')
data_engine2 = data_engine.drop(["game_id", "engine", "developer" ,"active_players", "avg_recent_playtime", "median_recent_playtime", "porting", "publisher", "mode", "perspective", "theme", "more_tags"], axis=1)
```

```python
def filterEmptyEngine(df):
    empty_indices2 = df[df['genre'] == ''].index
    df = df.drop(empty_indices2)
    return df

data_engine3 = filterEmptyEngine(data_engine2)
data_engine3
```

```python
data_engine3['release_date']
```

```python
data_engine3 = data_engine3.dropna(subset=['genre'])
data_engine3 = data_engine3.dropna(subset=['release_date'])
def parseDate(value):
    date = value.split(",")[-1].strip()
    date = re.sub(r'\.', '-', date)
    return date
    
data_engine3['release_date'] = data_engine3['release_date'].apply(parseDate)
```

```python
data_engine3
```

```python
data_engine3['month'] = pd.to_datetime(data_engine3['release_date'], format="mixed").dt.strftime('%Y')
data_engine4 = data_engine3.sort_values(by='month')
data_engine4.head(15)
```

```python
def process_engine_column(engine_column):
    processed_list = []
    for item in engine_column:
        item = re.sub(r'[\[\]]', '', item)
        item = re.sub(r'\'', '', item)
        item = re.sub(r'\s+', ' ', item) 
        item = item.split(',')
        processed_list.append([x.strip() for x in item])
    return processed_list

data_engine4['genre'] = process_engine_column(data_engine4['genre'])
```

```python
all_genres = []
for i,elem in enumerate(data_engine4['genre']):
    all_genres.extend(elem)

all_genres = list(set(all_genres))
```

```python
all_genres = [genre for genre in all_genres if genre and not re.search(r'&', genre)]
all_genres = sorted(all_genres)
all_genres 
```

```python
data_engine4
```

```python
dataset = pd.DataFrame(columns=['month', 'nb_game', 'estimated_players'])
for genre in all_genres:
    dataset[genre] = 0

for i, row in data_engine4.iterrows():
    for g in row['genre']:
        if g in all_genres:
            new_row = pd.DataFrame([row['month']], columns=['month'])
            new_row[g] = row['estimated_players']
            dataset = pd.concat([dataset, new_row], ignore_index=True)
```

```python
bar_dict = []
for genre in all_genres:
    bar_dict.append({"genre": genre, "nb_game": np.zeros(2025-1970+1), "nb_hour_played": np.zeros(2025-1970+1) })

bar_dict = {elem["genre"]: elem for elem in bar_dict}
```

```python
game_dataset_ordered = data_engine4#.sort_values(by="release_date", ascending=True)

def intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))

for i in range(len(game_dataset_ordered)):
    game = game_dataset_ordered.iloc[i]
    for g in game["genre"]:
        if g in all_genres:
            bar_dict[g]["nb_game"][int(game["month"])-1970] += 1
            bar_dict[g]["nb_hour_played"][int(game["month"])-1970] += game["avg_total_playtime"] 

bar_dict
        
```
