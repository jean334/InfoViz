```python
cleaned_games_dataset = pd.read_csv('dataset/games_dropped_column.csv')
```

```python
estimated_owners = cleaned_games_dataset["Estimated owners"].unique()
estimated_owners_map = {}
for i in range(len(estimated_owners)):
    estimated_owners_map[estimated_owners[i]] = i

cleaned_games_dataset["Estimated owners"] = cleaned_games_dataset["Estimated owners"].map(estimated_owners_map)

def process_genres(value):
    if isinstance(value, float):
        return []
    genres = value.split(",")
    genres = [genre.strip() for genre in genres]
    return genres

cleaned_games_dataset["genres"] = cleaned_games_dataset["genres"].apply(process_genres)
```

```python
def getPublisherGames(publisher):
    return cleaned_games_dataset[cleaned_games_dataset["Publishers"] == publisher]["name"]

publisher_list = cleaned_games_dataset["Publishers"].unique()
publisher_games = {}
for publisher in publisher_list:
    publisher_games[publisher] = getPublisherGames(publisher)

publisher_games
```

```python
def filterPublisher():
    new_publisher_games = {}
    number_of_publisher = 0
    for elem in publisher_games:
        #if len(publisher_games[elem]) > 4 and len(publisher_games[elem]) < 20 and number_of_publisher < 20:
        new_publisher_games[elem] = publisher_games[elem]
        number_of_publisher += 1
    return new_publisher_games

new_publisher_games = filterPublisher()
```

```python
def getAllPairs(publisher_games):
    all_pairs = []
    for key in publisher_games:
        if len(publisher_games[key]) >= 2:
            game_list = publisher_games[key].to_list()
            for i in range(len(game_list)):
                for j in range(i+1, len(game_list)):
                    all_pairs.append((game_list[i], game_list[j]))
            
    return all_pairs

all_pairs = getAllPairs(new_publisher_games)
```

```python
test = {"nodes": [{"id": name, "release_date": release_date,"estimated_owners": estimated_owners, "user_score": user_score, "average_playtime_forever": average_playtime_forever, "median_playtime_forever": median_playtime_forever, "publisher": publisher, "genres": genres} for name, release_date, estimated_owners, user_score, average_playtime_forever, median_playtime_forever, publisher, genres in zip(cleaned_games_dataset["name"], cleaned_games_dataset["release_date"], cleaned_games_dataset["Estimated owners"], cleaned_games_dataset["user_score"], cleaned_games_dataset["average_playtime_forever"], cleaned_games_dataset["median_playtime_forever"], cleaned_games_dataset["Publishers"], cleaned_games_dataset["genres"])]}
test["links"] = [{"source": name1, "target": name2, "source_group": [],"target_group": [], "value": 1} for name1, name2 in all_pairs]
```

## Create a small dataset

```python
def getListOfGames(publisher_games):
    game_list = []
    for elem in publisher_games:
        game_list.extend(publisher_games[elem].to_list())
    return game_list

game_list = getListOfGames(new_publisher_games)
print(game_list)


def filterNodes(dico, game_list):
    new_dico = []
    for elem in dico:
        if elem["id"] in game_list:
            new_dico.append( elem)
    return new_dico

new_dico = filterNodes(test["nodes"], game_list)
new_test = {"nodes": new_dico, "links": test["links"]}
```

```python
def getGroupNodes(game_name):
    for elem in new_test["nodes"]:
        if elem["id"] == game_name and len(elem["genres"]) > 0:
            return elem["genres"][0]

def getGroupLinks():
    for elem in new_test["links"]:
        elem["source_group"].append(getGroupNodes(elem["source"]))
        elem["target_group"].append(getGroupNodes(elem["target"]))
    

getGroupLinks()
```
