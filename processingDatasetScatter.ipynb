```python
import pandas as pd
import numpy as np
```

```python
game_dataset = pd.read_csv('dataset/games.csv')
game_dataset = game_dataset.sample(n=4000, random_state=1)
```

```python
game_dataset.rename({"": "AppID", "AppID": "Name", "Name": "Release date", "Release date": "Estimated owners", "Estimated owners": "Peak CCU", "Peak CCU": "Required age", "Required age": "Price", "Price": "osef1", "DiscountDLC count": "osef2"},axis=1, inplace=True)
```

```python
game_dataset.head(20)
```

```python
game_dataset2 = game_dataset.drop(['Release date', 'Peak CCU', 'Required age', 'osef1', 'osef2', 'About the game', 'Supported languages',
                            'Full audio languages', 'Reviews', 'Header image', 'Website', 'Support url', 'Support email', 'Windows',
                            'Mac', 'Linux', 'Metacritic score', 'Metacritic url', 'Achievements', 'Average playtime two weeks',
                            'Average playtime forever', 'Median playtime two weeks', 'Developers', 'Publishers', 'Tags', 'Screenshots', 'Movies'], axis=1)
```

```python
game_dataset3 = game_dataset2.drop(['Notes', 'Categories', 'Score rank'], axis=1)
```

```python
#filter out games with no recommendations and no Estimated owners

def filter_games(dataframe):
    filtered_df = dataframe[(dataframe['Recommendations'] != 0) & (dataframe['Estimated owners'] != 0)]
    return filtered_df

filtered_game_dataset = filter_games(game_dataset3)
filtered_game_dataset
```

```python
filtered_game_dataset.iloc[[1]]['Estimated owners']
```

```python
def format_estimated_owners(value):
    val = value.split(' ')
    val1 = val[0]
    val2 = val[2]
    return int((int(val1)+int(val2))/2)

filtered_game_dataset['Estimated owners'] = filtered_game_dataset['Estimated owners'].apply(format_estimated_owners)
```
