```python
import numpy as np
import pandas as pd
import re
```

```python
data_engine = pd.read_csv('playtracker_data_engine.csv')
data_engine2 = data_engine.drop(["game_id", "active_players", "avg_total_playtime", "avg_recent_playtime", "median_recent_playtime", "porting", "publisher", "mode", "perspective", "theme", "more_tags"], axis=1)
```

```python
data_engine2
```

```python
def filterEmptyEngine(df):
    empty_indices = df[df['engine'] == '[]'].index
    df = df.drop(empty_indices)
    empty_indices2 = df[df['genre'] == ''].index
    df = df.drop(empty_indices2)
    return df

data_engine3 = filterEmptyEngine(data_engine2)
data_engine3
```

```python
def process_engine_column(engine_column):
    processed_list = []
    for item in engine_column:
        # Remove square brackets and split by comma
        item = re.sub(r'[\[\]]', '', item)
        item = re.sub(r'\'', '', item)
        item = re.sub(r'\s+', ' ', item)  # Remove extra spaces
        item = item.split(',')
        processed_list.append([x.strip() for x in item])
    return processed_list

data_engine3['engine'] = process_engine_column(data_engine3['engine'])
data_engine3['genre'] = process_engine_column(data_engine3['genre'])
data_engine3
```

```python
def processEngine(value):
        for elem in value:
            if re.search("Unreal Engine [0-9]", elem):
                return "Unreal Engine"
            elif re.search("Source [0-9]", elem):
                return "Source"
            elif re.search("Halo .*", elem):
                return "Halo"
            elif re.search("id Tech [0-9]", elem):
                return "id Tech"
            elif re.search("Unity [0-9]", elem):
                return "Unity"
            elif re.search("LithTech .*", elem):
                return "LithTech"
            elif re.search("CryEngine [0-9]", elem):
                return "CryEngine"
            elif re.search("Glacier [0-9]", elem):
                return "Glacier"
            elif re.search("RPG Maker .*", elem):
                return "RPG Maker"
            else:
                return elem

data_engine3['engine'] = data_engine3['engine'].apply(processEngine)
data_engine3
```

```python
#remove NaN values from the engine column
print(len(data_engine3))
data_engine3 = data_engine3.dropna(subset=['engine'])
data_engine3 = data_engine3.dropna(subset=['genre'])
data_engine3 = data_engine3.dropna(subset=['release_date'])
len(data_engine3)
```

```python
# min and max date, binning date to month, counting engine per genre and engine for each half year
def parseDate(value):
    date = value.split(",")[-1].strip()
    date = re.sub(r'\.', '-', date)
    return date
    

data_engine3['release_date'] = data_engine3['release_date'].apply(parseDate)
data_engine3['month'] = pd.to_datetime(data_engine3['release_date']).dt.strftime('%Y/%m')
data_engine3
```

```python
data_engine4 = data_engine3.sort_values(by='month')
data_engine4.head(15)
```

```python
#get every genre in the dataset
all_genres = []
for i,elem in enumerate(data_engine4['genre']):
    for g in elem:
        if g != '':
            all_genres.extend(elem)

all_genres = list(set(all_genres))
```

```python
dataset = pd.DataFrame(columns=['month', 'engine'])
for genre in all_genres:
    dataset[genre] = 0

for i, row in data_engine4.iterrows():
    for g in row['genre']:
        if g != '':
            new_row = pd.DataFrame([[row['month'], row['engine']]], columns=['month', 'engine'])
            new_row[g] = row['estimated_players']
            dataset = pd.concat([dataset, new_row], ignore_index=True)
            """
            print("row['estimated_players']", row['estimated_players'])
            dataset.at[i, g] += row['estimated_players']
            dataset.at[i, 'engine'] = row['engine']
            dataset.at[i, 'month'] = row['month']
            """
```

```python
#replace NaN values with 0
dataset = dataset.fillna(0)
```

```python
dataset = dataset.groupby(['month', 'engine']).sum().reset_index()
dataset
```

```python
dataset.to_csv('my-app/public/assets/stacked_engine_3.csv', index=False)
```
